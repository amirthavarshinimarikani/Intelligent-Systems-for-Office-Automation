```python
# a. Import the necessary libraries
import pandas as pd
import re
from textblob import TextBlob
from google.colab import drive

df = pd.read_csv('/content/drive/MyDrive/IMDB Dataset.csv')

print("âœ… Dataset Loaded Successfully!")
display(df.head())

texts = df['review']

def remove_handles(text):
    return re.sub(r'@\w+', '', text)

clean_texts = texts.apply(remove_handles)

def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity
    if polarity > 0:
        return 'positive'
    elif polarity < 0:
        return 'negative'
    else:
        return 'neutral'

predicted_sentiments = clean_texts.apply(get_sentiment)

sentiment_df = pd.DataFrame({'clean_text': clean_texts, 'predicted_sentiment': predicted_sentiments})

merged_df = pd.concat([df, sentiment_df], axis=1)

positive_reviews = merged_df[merged_df['predicted_sentiment'] == 'positive']
negative_reviews = merged_df[merged_df['predicted_sentiment'] == 'negative']
neutral_reviews  = merged_df[merged_df['predicted_sentiment'] == 'neutral']

print("\nâœ… Positive Reviews:")
display(positive_reviews.head())

print("\nðŸš« Negative Reviews:")
display(negative_reviews.head())

print("\nðŸ˜ Neutral Reviews:")
display(neutral_reviews.head())

print("\nðŸ“Š Sentiment Count Summary:")
print(merged_df['predicted_sentiment'].value_counts())
```
